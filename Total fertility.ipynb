```python
import csv
from matplotlib import pyplot as plt 

country = []
firstYear = []
secondYear = []
width = 0.8

with open('TFs.csv') as file:
    reader = csv.DictReader(file)
   
    for row in reader:
        country.append(row['Location'])
        firstYear.append(row['2010 - 2015'])
        secondYear.append(row['2015 - 2020'])
firstYear = [float(num) for num in firstYear]
secondYear = [float(num) for num in secondYear]
plt.figure(figsize = (20, 100))
plt.barh(country,firstYear, 0.8, label='2010 - 2015')
plt.barh(country, secondYear, 0.8, left= firstYear, label='2015 - 2020')
plt.ylabel('Countries', fontsize=30)
plt.xlabel('Total Fertility', fontsize=30)
plt.tick_params(length= 30 , pad=10)
plt.yticks(rotation=0, fontsize=30)
plt.xticks(rotation=45, fontsize=30)
plt.legend(fontsize = 30,)
plt.grid()
plt.axis([0,15, 0, __])


plt.show()
```

```python
import math

# Y-axis values 
firstYear = [float(i) for i in firstYear]
secondYear = [float(i) for i in secondYear]

sqs =[]
# standard deviation

# adding all the number in the array
ys= sum(firstYear)

# getting the mean
m = ys / len(firstYear)

# looping thru the array 
for i in firstYear:
    # adding each number in the array with the mean 
    sq = (i - m) ** 2
    sqs.append(sq)

# summation of the numerator
n = sum(sqs)

# 
z = n/ (len(firstYear) - 1) 

# Standard deviation 
SD = math.sqrt(z)
print(f'standard deviation = {SD:.2f}')


### SECOND YEAR ###
sqs =[]
# standard deviation

# adding all the number in the array
ys= sum(secondYear)

# getting the mean
m = ys / len(secondYear)

# looping thru the array 
for i in secondYear:
    # adding each number in the array with the mean 
    sq = (i - m) ** 2
    sqs.append(sq)

# summation of the numerator
n = sum(sqs)

# 
z = n/ (len(firstYear) - 1) 

# Standard deviation 
SD = math.sqrt(z)
print(f'standard deviation = {SD:.2f}')
```
